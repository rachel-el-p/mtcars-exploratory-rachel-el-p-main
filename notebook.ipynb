# Exploratory data analysis

So far we've done a lot of looking at distributions of single variables. In this dataset we have some continuous variables (like `mpg`, `disp`, `hp`, `drat`, `wt`, and `qsec`) and some discrete variables (like `vs`, `am`, `gear`, and `carb`). Let's quickly glance through the distributions of them.




```python
%%R 

continuous_variables <- c('mpg', 'disp', 'hp', 'drat', 'wt', 'qsec')
discrete_variables   <- c('vs', 'am', 'gear', 'carb')

# Conevert to long format
mtcars_long = mtcars %>% pivot_longer(!name, names_to = 'variable', values_to = 'value') 
mtcars_long

#plot only continuous variables
to_plot <- mtcars_long %>% filter(variable %in% continuous_variables)
to_plot

histogram_means <- to_plot %>%
group_by(variable) %>%
summarise(mean = mean(value))
histogram_means

plot_continuous_histograms <- ggplot(to_plot) +
    aes(x=value, fill=variable, group=variable) +
    geom_histogram(bins=5) +
    geom_vline(data=histogram_means, aes(xintercept=mean)) +
    facet_wrap(~variable, scales='free') +
    theme_minimal() +
    labs(title = 'mtcars at a glance', subtitle='continuous variables (means marked with a vertical line)')
plot_continuous_histograms


plot_continuous_dotplot <- ggplot(to_plot) +
    aes(x=value, y=1, color=variable) +
    geom_point(size=4, alpha=.2) +
    stat_summary(fun.y= mean, geom="point", shape='x',
                 size=10, color="black", alpha=1, orientation='y') +
    facet_wrap(~variable, scales='free', ncol=1) +
    theme_minimal() +
    labs(
        title = 'mtcars at a glance', 
        subtitle='continuous variables (means marked with an X)'
    ) +
    theme(
            axis.text.y=element_blank(),  #remove y axis labels
            axis.ticks.y=element_blank(),  #remove y axis ticks
            panel.grid.major.y = element_blank(), #remove y axis grid
            panel.grid.minor.y = element_blank(), #remove y axis grid
    )

plot_continuous_dotplot

# # display both plots
print(plot_continuous_histograms)
print(plot_continuous_dotplot)
```

```python
%%R 

# plot only discrete variables
to_plot <- mtcars_long %>% filter(variable %in% discrete_variables)

plot_discrete <- ggplot(to_plot) +
    aes(y=value, fill=variable) +
    geom_bar() +
    facet_wrap(~variable, scales='free') +
    theme_minimal() +
    labs(title = 'mtcars at a glance', subtitle='discrete variables')


plot_discrete
```

## Part 1 (Gudied Inquiry)
#  What makes cars more fuel efficient?


```python
%%R -w 1000

ggplot(mtcars) +
    aes(y=mpg, x=qsec, color='red', size=1, alpha=0.5) +
    geom_point() +
    theme_minimal() +
    labs(
        title = 'Want to save money on gas? You may have to drive slower', 
        subtitle='Cars that use less fuel also take longer to hit 1/4 mile'
    ) +
    xlab("Fastest time to travel 1/4 mile from standstill (in seconds)") + ylab("miles per gallon")+
    theme(
            plot.title = element_text(size=22))
```

```python
%%R -w 500 -h 400
ggplot(mtcars)+ 
    aes(y=mpg, x=gear) +
    geom_col(fill='black') +
    labs(
        title = 'Cars with more gears are less fuel efficient', 
        subtitle='Cars with five gears 
can only go up to 100 miles before a gallon of gas is gone. Those with three
or four gears can go at least twice as long.'
    ) +
    xlab("number of gears") + ylab("miles per gallon")+
    theme(
            plot.title = element_text(size=16))
```

```python
%%R -w 500 -h 400
ggplot(mtcars)+ 
    aes(y=mpg, x=am)+
    geom_col(fill='black')+
    labs(
        title = "Fuel efficiency doesn't depend on transmission type" , 
        subtitle= "Whether a car is automatic (am=0) or manual (am=1) doesn't affect how much fuel
a car uses."
    ) +
    xlab("transmission configuration") + ylab("miles per gallon")+
    theme(
            plot.title = element_text(size=16))
```

```python
%%R -w 1000

ggplot(mtcars) +
    aes(y=mpg, x=hp, color='red', size=1, alpha=0.5) +
    geom_point() +
    geom_quantile(colour = "blue", linewidth = 1, alpha = 0.5, quantiles = 0.5) +
    theme_minimal() +
    labs(
        title = 'Cars with less horsepower are more fuel efficient', 
        subtitle='Cars with less than 100hp can go between 20 to 35 miles per gallon.
Cars with more than 100hp can go between 10 and 20 miles per gallon.'
    ) +
    xlab("Engines power output('gross horsepower')") + ylab("miles per gallon")+
    theme(
            plot.title = element_text(size=22))
```

```python
%%R -w 1000

ggplot(mtcars) + 
    aes(x=qsec, y='carb') +
    geom_point(size=5,alpha=.5,outlier.size=10) + 
    stat_summary(geom='point',fun.y='mean',shape='x',size=10, color="red", alpha=1) +
    facet_wrap(~ carb) + 
    theme_minimal() +
    labs(
        title = 'Cars with more carburetors have less horsepower', 
        subtitle='Cars with three to eight carburetors had a horsepower of less than 18, which those with
one or two had between 18 and 20 horsepower.'
    ) +
    xlab("Engines power output('gross horsepower')") +
    theme(
            plot.title = element_text(size=22))
```
